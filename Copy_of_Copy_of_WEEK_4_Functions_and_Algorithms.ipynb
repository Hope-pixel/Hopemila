```python

def compute_average(scores):
    total = 0

    for score in scores:
        total = total + score

    average = total / len(scores)

    return average


scores = [85,30,70,45,60]

avg = compute_average(scores)

print("Average score:", avg)
```

```python

def check_pass_fail(score):

    if score >= 50:
        return "Pass"
    else:
        return "Fail"


print(check_pass_fail(85))
print(check_pass_fail(30))
```

```python

scores = [85,30,70,45,60]

for score in scores:
    result = check_pass_fail(score)
    print(score, result)
```
