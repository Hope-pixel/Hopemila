DATA SET OF 15 STUDENTS

```python

students = ["Ann", "Shelly","Grace","Mary","Ella","Mila","Judge","Caleb","Lisa","Mercy","Hope","Judith","Nancy","Josh","Joy"]
math_scores = [88,27,63,26,38,65,35,45,60,90,75,25,30,85,50]
english_scores = [56,92,70,78,67,35,85,70,25,35,45,68,65,40,80]
science_scores = [66,44,54,81,42,55,90,70,40,65,80,50,45,30,25]
```

HISTOGRAM (MATHS)

```python

plt.figure()
plt.hist(math_scores, bins=5)
plt.title("Distribution of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

HISTOGRAM (ENGLISH)

```python

plt.figure()
plt.hist(english_scores, bins=5)
plt.title("Distribution of English Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

HISTOGRAM (SCIENCE)

```python

plt.figure()
plt.hist(science_scores, bins=5)
plt.title("Distribution of science Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

SCATTER PLOT(MATHS VS ENGLISH)

```python

plt.figure()
plt.scatter(math_scores, english_scores)
plt.title("Math vs English Scores")
plt.xlabel("Math Scores")
plt.ylabel("English Scores")
plt.show()
```

SCATTER PLOT(ENGLISH VS SCIENCE)

```python

plt.figure()
plt.scatter(english_scores, science_scores)
plt.title("english vs science Scores")
plt.xlabel("english Scores")
plt.ylabel("science Scores")
plt.show()
```

SCATTER PLOT (SCIENCE VS MATHS)

```python

plt.figure()
plt.scatter(science_scores, math_scores)
plt.title("science vs math Scores")
plt.xlabel("science Scores")
plt.ylabel("Math Scores")
plt.show()
```

MULTIPLE LINE CHART(MATH & ENGLISH)
