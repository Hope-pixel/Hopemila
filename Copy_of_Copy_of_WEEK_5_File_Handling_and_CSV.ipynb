1. Read the CSV file.
2. Compute:
   - Average Math score
   - Average English score
   - Average Science score
3. Find:
   - Highest overall score
   - Lowest overall score
4. Print all results clearly.

Submit your completed notebook (.ipynb)


```python

file = open("Copy of week5_student_scores.csv")

next(file)

total = 0
count = 0

for line in file:
    data = line.strip().split(",")
    math_score = int(data[2])

    total += math_score
    count += 1

average = total / count

print("Average Math Score:", average)

file.close()
```
