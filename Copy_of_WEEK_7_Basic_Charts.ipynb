LINE CHARTS (MATHS SCORES)

```python

plt.figure()
plt.plot(students, math_scores)
plt.title("Math Scores - Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

LINE CHARTS (ENGLISH SCORES)

```python

plt.figure()
plt.plot(students, english_scores)
plt.title("english Scores - Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

LINE CHARTS (SCIENCE SCORES)

```python

plt.figure()
plt.plot(students, science_scores)
plt.title("science Scores - Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

BAR CHART (MATHS SCORES)

```python

plt.figure()
plt.bar(students, math_scores)
plt.title("Math Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

BAR CHART (ENGLISH SCORES)

```python

plt.figure()
plt.bar(students, english_scores)
plt.title("english Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

BAR CHART (SCIENCE SCORES)

```python

plt.figure()
plt.bar(students, science_scores)
plt.title("science Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

PIE CHARTS (MATHS SCORES)

```python

plt.figure()
plt.pie(math_scores, labels=students, autopct='%1.1f%%')
plt.title("Math Score Distribution")
plt.show()
```

PIE CHARTS (ENGLISH SCORES)
