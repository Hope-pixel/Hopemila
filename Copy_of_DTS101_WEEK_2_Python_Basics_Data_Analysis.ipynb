```python
# open the csv file for the python script to read
file = open('Copy of DTS101_week2_student_scores.csv')

# initial variables
total = 0
count = 1

# The next() function in Python retrieves the next item from an iterator
# for now, think of this line as us skipping the heading of our csv file
# (that contains student_id,math,english,physics,chemistry)
next(file)

# loop through the opened file
for line in file:
    # split the lines into different parts using the ',' as a delimiter
    parts = line.strip().split(',')
    # remember the english is at position 0
    english_score = int(parts[1])
    # calulate the total for english score
    total += english_score
    # count the line
    count += 1

# calculate and print the average for the english score
average = total / count
print('Average english score:', average)

# close the file
file.close()
```
